```python
def json_file_creator(record, file_name):
  json_ob = json.dumps(record, indent=4)
  with open(f"{file_name}.json", "w") as json_file:
    json_file.write(json_ob)
```

```python
 def sales_menu():
  global sales_count
  while True:
    try:
      print("Enter The Product ID: ")
      product_id = int(input())
      if product_id in record.keys():
        print("Enter The Number Of Units To Sell: ")
        units = int(input())
        if record[product_id]['Units'] - units >= 0:
          record[product_id]['Units'] -= units
          sales_count += 1
          sales[sales_count] = {'Produnt ID': product_id, 'Product': record[product_id], "Amount": record[product_id]['Price'] * units}
          print("---------------------------------------------------Bill--------------------------------------------------------")
          print(f"Product ID: {product_id} | Name: {record[product_id]['Name']} | Amount({units} x {record[product_id]['Price']}): {record[product_id]['Price'] * units}".center(108))
          print("------------------------------------------------Thank You-------------------------------------------------------")
          json_file_creator(record, 'record') # Updating The JSON File
          print()
          break
        else:
          print(f"Sorry, {units} units of {record[product_id]['Name']} is not available. Try Again")
          continue
      else:
        print(f"Sorry, no product with given product id is present in our inventory")
        continue
    except (ValueError, TypeError):
      print("Enter The Correct Product ID")
      continue
```

```python
def add_to_inventory():
  while True:
    try:
      print("Enter Product ID: ")
      product_id = int(input())
      if product_id in record.keys():
        print("Enter The Number Of Units Of Units To Add: ")
        units_to_add = int(input())
        record[product_id]['Units'] += units_to_add
        print("Success!")
        json_file_creator(record, 'record')
        break
      else:
        print("Enter the Name Of Product: ")
        name = input()
        print("Enter RAM Size: ")
        ram = int(input())
        print("Enter ROM Size:")
        rom = int(input())
        print("Enter Main Camera Size: ")
        camera = int(input())
        print('Enter the number of units:')
        units_to_add = int(input())
        record[product_id] = {'Name': name, 'RAM': ram, 'ROM': rom, 'Main Camera': camera, "Units": units_to_add}
        json_file_creator(record, 'record')
        print("Success!")
        break
    except (ValueError, TypeError):
      print("Enter the values correctly")
      continue
```

```python
def sales_report():
  total = 0
  for y in sales.values():
    total += y['Amount']
  sales['Total Sale Amount'] = total
  json_file_creator(sales, 'sales')
  print("Sales Report Of The Day Is Successfully Generated")
  print()
```

```python
def main_menu():
  while True:
    try:
      print("----------------------------------------Welcome to ETG SHOP For Smartphones-------------------------------------")
      print("You can:\n1.Display Products\n2.Sell Product\n3.Add New Product To Inventory\n4.Create Sales Report\n5.Exit\nChoose Anyone Option To Proceed...")
      option = int(input())
      if 1 <= option <= 5:
        if option == 1:
          display_menu()
        elif option == 2:
          sales_menu()
        elif option == 3:
          add_to_inventory()
        elif option == 4:
          sales_report()
        elif option == 5:
          break
      else:
        print("Choose Valid Option1")
        continue
    except (ValueError, TypeError):
      print("Choose Valid Option")
      continue
```
